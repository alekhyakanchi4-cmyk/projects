```python
students = []

while True:
    print("\n**************")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Exit")
    print("**************")

    ch = int(input("Enter choice: "))

    if ch == 1:
        roll = int(input("Enter Roll No: "))
        name = input("Enter Name: ")
        branch = input("Enter Branch: ")
        marks = float(input("Enter Marks: "))

        student = [roll, name, branch, marks]
        students.append(student)

        print("Student Added Successfully")

    elif ch == 2:
        if len(students) == 0:
            print("No Students Available")
        else:
            print("\nStudent Records")
            for i in students:
                print(i)

    elif ch == 3:
        r = int(input("Enter Roll No: "))

        for i in students:
            if i[0] == r:
                print("\nStudent Found")
                print(i)
                break
        else:
            print("Student Not Found")

    elif ch == 4:
        r = int(input("Enter Roll No: "))

        for i in students:
            if i[0] == r:
                i[1] = input("Enter New Name: ")
                i[2] = input("Enter New Branch: ")
                i[3] = float(input("Enter New Marks: "))

                print("Student Updated Successfully")
                break
        else:
            print("Student Not Found")

    elif ch == 5:
        r = int(input("Enter Roll No: "))

        for i in students:
            if i[0] == r:
                students.remove(i)
                print("Student Deleted Successfully")
                break
        else:
            print("Student Not Found")

    elif ch == 6:
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```

```python
students = []

def add_student():
    roll = int(input("Enter Roll No: "))
    name = input("Enter Name: ")
    branch = input("Enter Branch: ")
    marks = float(input("Enter Marks: "))

    student = [roll, name, branch, marks]
    students.append(student)

    print("Student Added Successfully")


def view_students():
    if len(students) == 0:
        print("No Students Available")

    else:
        print("\nStudent Records")
        for i in students:
            print(i)


def search_student():
    r = int(input("Enter Roll No: "))

    for i in students:
        if i[0] == r:
            print("\nStudent Found")
            print(i)
            break

    else:
        print("Student Not Found")


def update_student():
    r = int(input("Enter Roll No: "))

    for i in students:
        if i[0] == r:

            i[1] = input("Enter New Name: ")
            i[2] = input("Enter New Branch: ")
            i[3] = float(input("Enter New Marks: "))

            print("Student Updated Successfully")
            break

    else:
        print("Student Not Found")


def delete_student():
    r = int(input("Enter Roll No: "))

    for i in students:
        if i[0] == r:
            students.remove(i)

            print("Student Deleted Successfully")
            break

    else:
        print("Student Not Found")


while True:

    print("\n**************")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Exit")
    print("**************")

    ch = int(input("Enter Choice: "))

    if ch == 1:
        add_student()

    elif ch == 2:
        view_students()

    elif ch == 3:
        search_student()

    elif ch == 4:
        update_student()

    elif ch == 5:
        delete_student()

    elif ch == 6:
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```

```python
students = []


def load_students():
    try:
        with open("students.txt", "r") as f:
            for line in f:
                data = line.strip().split(",")

                students.append([
                    int(data[0]),
                    data[1],
                    data[2],
                    float(data[3])
                ])

    except FileNotFoundError:
        pass


def save_students():

    with open("students.txt", "w") as f:

        for i in students:

            f.write(f"{i[0]},{i[1]},{i[2]},{i[3]}\n")


def add_student():

    roll = int(input("Roll No : "))

    name = input("Name : ")

    branch = input("Branch : ")

    marks = float(input("Marks : "))

    students.append([roll, name, branch, marks])

    save_students()

    print("Student Added Successfully")


def view_students():

    if len(students) == 0:
        print("No Students")

    else:

        for i in students:
            print(i)


def search_student():

    r = int(input("Roll No : "))

    for i in students:

        if i[0] == r:
            print(i)
            break

    else:

        print("Student Not Found")


def update_student():

    r = int(input("Roll No : "))

    for i in students:

        if i[0] == r:

            i[1] = input("New Name : ")

            i[2] = input("New Branch : ")

            i[3] = float(input("New Marks : "))

            save_students()

            print("Updated")

            break

    else:

        print("Student Not Found")


def delete_student():

    r = int(input("Roll No : "))

    for i in students:

        if i[0] == r:

            students.remove(i)

            save_students()

            print("Deleted")

            break

    else:

        print("Student Not Found")


load_students()


while True:

    print("\nStudent Management System")

    print("1 Add")

    print("2 View")

    print("3 Search")

    print("4 Update")

    print("5 Delete")

    print("6 Exit")


    ch = int(input("Choice : "))


    if ch == 1:
        add_student()


    elif ch == 2:
        view_students()


    elif ch == 3:
        search_student()


    elif ch == 4:
        update_student()


    elif ch == 5:
        delete_student()


    elif ch == 6:

        print("Thank You")

        break


    else:

        print("Invalid Choice")
```
